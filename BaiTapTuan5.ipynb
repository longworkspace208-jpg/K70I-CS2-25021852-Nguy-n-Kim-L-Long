```python
#W5A4

def divisorCounter(x):
    res = 0
    for i in range (1, int(x ** 0.5) + 1):
        if x % i == 0:
            res += i
            if (x / i != i):
                res += x / i
    return res

x = int(input())
if divisorCounter(x) == x:
    print("So hoan hao")
else:
    print("So khong hoan hao")


```

```python

#W5A6
def GiaiThua(x):
    if (x == 1):
        return 1
    return x * GiaiThua(x - 1)
n = int(input())
print(GiaiThua(n))
```

```python
#W5A7
def Cal(x, y, ty):
    match ty:
        case "+":
            print(x + y)
        case "-":
            print(x - y)
        case "*":
            print(x * y)
        case "/":
            print(f"{(x / y):.2f}")
        case _:
            print("Syntan error")
x, y = map(int, input().split())
ty = input()
Cal(x, y, ty)
```

```python
def HammingCounter(x, y):
    res = 0
    while (x != 0 or y != 0):
        res += (x % 2 != y % 2)
        x //= 2
        y //= 2
    return res

x, y = map(int, input().split())
print(HammingCounter(x, y))
```

```python
def Sumdig(x):
    res = 0
    while (x > 0):
        res += x % 10
        x //= 10
n = int(input())
print(Sumdig(n))
```
