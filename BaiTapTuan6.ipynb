```python
#W6A1
a = list(map(int, input().split()))
d = {}
print('[', end = '')
for index,value in enumerate(a):
    if d.get(value) == None:
        d[value] = 1
        if (index == 0):
            print(f"{value}", end = '')
        else:
            print(", ", end = '')
            print(value, end = '')
    else:
        del a[index]
print(']')
```

```python
#W6A2

a = list(map(int, input().split()))
b = [0] * (len(a) + 5)

b[0] = a[0]
for i in range(1, len(a)):
    b[i] = b[i - 1] + a[i]
for i in range(0, len(a)):
    print(b[i], end = ' ')
```

```python
#W6A4
s = input().split()
d = {}
for i in s:
    x, y = i.split(':')
    if (d.get(x) == None):
        d[x] = [y]
    else:
        d[x].append(y)
for index, (key, value) in enumerate(d.items()):
    print(f"{key} : {value}")
    
    

```

```python
#W6A5
d = {
    "positives":0,
    "negatives":0,
    "zeros":0    
}
n = list(map(int, input().split()))
for i in n:
    if i > 0:
        d["positives"] += 1
    elif i == 0:
        d["zeros"] += 1
    else:
        d["negatives"] += 1
for key, value in d.items():
    print(f"{key} : {value}")
```

```python
my_list = list(map(int, input().split()))
k = int(input())

output_list = []
d = {}
for i in my_list:
    if d.get(i) == None:
        d[i] = 1
    else:
        d[i] += 1
for i in my_list:
    if d.get(k - i) != None:
        output_list.append((i, k - i))
print(output_list)
```

```python
d = {}
my_list = map(int, input().split())
for i in my_list:
    if d.get(i) == None:
        d[i] = 1
    else:
        d[i] += 1
ans, Counter = 0, 0
for key, value in d.items():
    if (Counter < value):
        ans, Counter = key, value
print (ans) 
```
