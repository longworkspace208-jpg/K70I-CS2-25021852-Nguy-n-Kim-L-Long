```python
#W4A8

def check(x):
    if (x[:len(x) // 2:1] == x[:round(len(x) / 2) - 1:-1]):
        return True
    return False

def Solve(x, step):
    if check(str(x)) == True:
        print(x)
        print(step)
        return
    revx = str(x)[::-1]
    Solve(x + int(revx), step + 1)

n = int(input())
Solve(n, 1)
```

```python
#W4A9 PerfectSquaresDistinctDigits

def check(x):
    a = [0] * 10
    while (x > 0):
        a[x % 10] += 1
        x //= 10
    for i in a:
        if i >= 2:
            return False
    return True

n = int(input())
i, dem = 1, 0
while (i * i <= n):
    if (check(i * i) == True):
        dem += 1
        print(i * i, end = " ")
    i += 1
if (dem == 0):
    print("No number")
    
```

```python

#W4A10
def CollatzCounter(x):
    step = 0
    while (x != 1):
        if x % 2 == 0:
            x /= 2
        else:
            x = x * 3 + 1
        step += 1
    return step

n = int(input())
a = [0]
ans = 0
for i in range(1, n + 1):
    a.append(CollatzCounter(i))
    ans = max(ans, a[-1])
for index, value in enumerate(a):
    if value == ans:
        print(index, value)
        break
```

```python
#4A13

def divisorCounter(x):
    res = 0
    for i in range (1, int(x ** 0.5) + 1):
        if x % i == 0:
            res += i
            if (x / i != i):
                res += x / i
    return res

n = int(input())
for i in range (n):
    x, y = map(int, input().split())
    if divisorCounter(x) - x == y and divisorCounter(y) - y == x:
        print("YES")
    else:
        print("NO")


```
