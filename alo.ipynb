```python
l, r = map(int, input().split())

tmp = r // 10 - l // 10
ans = tmp * 4

if l % 10 != 0:
    for i in range(l, min(r, (l // 10 + 1) * 10 + 1)):
        if i % 10 == 1 or i % 10 == 4 or i % 10 == 6 or i %10 == 9:
            ans += 1
if r % 10 != 0:
    for i in range(l, max(l, (l // 10) * 10 + 1)):
        if i % 10 == 1 or i % 10 == 4 or i % 10 == 6 or i %10 == 9:
            ans += 1
print(ans)
```

```python
l, r = map(int, input().split())
cnt = 0
for i in range(l, r +  1):
    if (i ** 2 - 1) % 5 == 0:
        cnt += 1
print(cnt)



```

```python
#Cau 2
n = int(input())
for i in range(n):
    s = input()
    ans = 0
    cnt = 0
    for i in s:
        if i == 'A' or i == 'N' or i == 'H':
            ans = max(ans, cnt)
            cnt = 0
        else:
            cnt += 1
    if ans != 0:
        print(max(ans, cnt))
    else: 
        print(-1)


```
